Downloaded S&P 500 Data

```python
import yfinance as yf
sp500 = yf.Ticker("^GSPC")
sp500_data = sp500.history(period="max")
sp500_data
```

Clean and Visualize the Dataset

```python
sp500_data.index
```

```python
sp500_data.plot.line(y="Close", use_index=True)
```

```python
del sp500_data['Dividends']
del sp500_data['Stock Splits']

sp500_data["Tomorrow"] = sp500_data['Close'].shift(-1)
sp500_data
```

Set Up Machine Learning Target

```python
sp500_data['Target'] = (sp500_data['Tomorrow'] > sp500_data['Close']).astype(int)
sp500_data
```

```python
sp500_data = sp500_data.loc['1990-01-01':].copy()
sp500_data
```

Trained the Initial Model

The random forest classifier creates and limits the inaccuracy of each decision tree algorithm, and picks the best   

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
train = sp500_data.iloc[:-100]
test = sp500_data.iloc[-100:]

predictors = ['Close', 'Volume', 'Open', "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors]) ## makes the predictions of the random forest tree

import pandas as pd
preds = pd.Series(preds, index=test.index)
precision_score(test['Target'], preds)
```

```python
combined = pd.concat([test['Target'], preds], axis=1)
combined.plot()
```

Evaluated Error and Created Backtesting

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500_data, model, predictors)
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions['Target'].value_counts() / predictions.shape[0]
```

Additional Predictions to Improve the Model

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_avg = sp500_data.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500_data[ratio_column] = sp500_data["Close"] / rolling_avg['Close']

    trend_column = f"Trend_{horizon}"
    sp500_data[trend_column] = sp500_data.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]

sp500_data = sp500_data.dropna()
sp500_data
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined

predictions = backtest(sp500_data, model, new_predictors)
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
